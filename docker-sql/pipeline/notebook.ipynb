```python
import pandas as pd
```

```python
prefix = 'https://github.com/DataTalksClub/nyc-tlc-data/releases/download/yellow'
url = f'{prefix}/yellow_tripdata_2021-01.csv.gz'
url
```

```python
dtype = {
    "VendorID": "Int64",
    "passenger_count": "Int64",
    "trip_distance": "float64",
    "RatecodeID": "Int64",
    "store_and_fwd_flag": "string",
    "PULocationID": "Int64",
    "DOLocationID": "Int64",
    "payment_type": "Int64",
    "fare_amount": "float64",
    "extra": "float64",
    "mta_tax": "float64",
    "tip_amount": "float64",
    "tolls_amount": "float64",
    "improvement_surcharge": "float64",
    "total_amount": "float64",
    "congestion_surcharge": "float64"
}

parse_dates = [
    "tpep_pickup_datetime",
    "tpep_dropoff_datetime"
]

df = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates
)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5432/ny_taxi')
# syntax explain:
# postgresql://<username>:<password>@<host>:<port>/<database_name>
```

```python
print(pd.io.sql.get_schema(df, name='yellow_taxi_data', con=engine))
```

```python
# take the 0th row of df (i.e. just column names and dtypes)
# write them to the database as a table named 'yellow_taxi_data' using the engine connection
# drop the table if it already exists and recreates it
# end up with an empty table that matches the DataFrame schemas
df.head(n=0).to_sql(name='yellow_taxi_data', con=engine, if_exists='replace')
```

```python
df_iter = pd.read_csv(
    url,
    dtype=dtype,
    parse_dates=parse_dates,
    iterator=True,
    chunksize=100000
)
```

```python
!uv add tqdm
```

```python
from tqdm import tqdm
```

```python
first = True

for df_chunk in tqdm(df_iter, desc="Inserting chunks"):

    if first:
        # Create table schema (no data)
        df_chunk.head(0).to_sql(
            name="yellow_taxi_data",
            con=engine,
            if_exists="replace"
        )
        first = False
        print("Table created")

    # Insert chunk
    df_chunk.to_sql(
        name="yellow_taxi_data",
        con=engine,
        if_exists="append"
    )

    print("Inserted:", len(df_chunk))
```
